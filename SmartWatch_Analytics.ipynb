```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import plotly.graph_objects as go
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/dailyActivity_merged.csv")

data.head()
```

### Checking for null values

```python
data.isnull().sum()
```

>It appear there are no null values in the dataset

### Getting info about the dataset

```python
data.info()
```

The column ActivityDate is an object so we have to convert it into date type if we want to use this column in our analysis

```python
# converting ActivityDate to date 

data.ActivityDate = pd.to_datetime(data['ActivityDate'],format="%m/%d/%Y")
```

#### Getting the total minutes collated by the smartwatch on each day by adding the VeryActiveMinutes,FairlyActiveMinutes,LightlyActiveMinutes, SedentaryMinutes together

```python
data['TotalMinutes'] = data['VeryActiveMinutes'] + data['FairlyActiveMinutes'] + data['LightlyActiveMinutes'] + data['SedentaryMinutes']
```

### Getting a descriptive statistics of the dataset

```python
data.describe()
```

### Analyzing the Smartwatch Data

The dataset has a “Calories” column; it contains the data about the number of calories burned in a day. Let’s have a look at the relationship between calories burned and the total steps walked in a day:

```python
figure = px.scatter(data_frame = data, x="Calories",
                    y="TotalSteps", size="VeryActiveMinutes", 
                    trendline="ols", 
                    title="Relationship between Calories & Total Steps")

figure.show()
```

We can see there's a linear corroletion between total steps and calories burned as expected.


Now we can look how user activity is distributed on average

```python
label = ["Very Active Minutes", "Fairly Active Minutes", "Lightly Active Minutes", "Inactive Minutes"]

values = data[["VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "SedentaryMinutes"]].mean().round()

colors = ['gold','lightgreen', "pink", "blue"]

fig = go.Figure(data=[go.Pie(labels= label, values= values )])
fig.update_layout(title_text='Total Active Minutes')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```
